#### REGRESIÓN LOGÍSTICA CON SCIKIT-LEARN

Algoritmo de clasificación que se usa para predecir la probabilidad de una variable dependiente categórica. El modelo logit predice $P(Y=1)$ como una función de $X$. Se modela la probabilidad de una forma tal que los outputs serán valores entre 0 y 1 para cualquier valor de $X$.


Ahora utilizaremos la función [LogisticRegression()](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html)

Se pueden proveer muchos parámetros opcionales para esta función:

- **fit_intercept**: Boolean que decide si calcular el intercepto (True) o considerarlo igual a cero (False). Por default es True.
- **penalty**: Se determina se usar algún tipo de regularización (lo veremos mas adelante en el curso). Posibles valores: ‘l1’, ‘l2’, ‘elasticnet’ y 'None' (usaremos esta opción por ahora). El valor por defecto es default es ‘l2’, es decir que se aplica regularización.


```python
# Creamos un vector de x e y para fines del ejemplo.
np.random.seed(25)
X = np.random.normal(size=5000) #5000 obs
print(X)

# Recordatorio: para la regresión lineal creamos un vector aleatorio así:
y_int = 2 + 3*X + np.random.rand(500, 1)

# Ahora lo crearemos de la siguiente forma para que tenga más sentido usar una regresión logística
y = (X > 0).astype(float) # si no pusiera astype sería un array de True y False

# Alteramos los valores de X y sumamos variación con el "error"
X[X > 0] *= 25
X += .5 * np.random.normal(size=5000)
X = X.reshape((-1, 1)) # para tenerlo como columna
print(X)

```

```python
X.shape
```

```python
# Graficamos para ver si nos quedó un vector que 1s y 0s
plt.scatter(X, y, color='orange', zorder=20, marker="|")
plt.xlabel('x')
plt.ylabel('Y')
plt.show()
```

#### Estimación de Regresión logistica

```python
# Ajustamos el clasificador con el método fit() 
log_reg = LogisticRegression(penalty=None).fit(X, y) 

# Estimacion de los coeficientes
print(log_reg.coef_)
print(log_reg.intercept_)
```

Prediccion de probabilidad $P(Y=1|X)$

```python
# Predicciones (probabilidad) (p sobrerito)
y_pred_score = log_reg.predict_proba(X)[:,1]  
print('promedio de predicción Prob(y=1|X): ', y_pred_score.mean().round(3))
print('min de predicción Prob(y=1|X): ', y_pred_score.min().round(3))
print('max de predicción Prob(y=1|X): ', y_pred_score.max().round(3))
```

Efectivamente nuestra predicción de la probabilidad $P(Y=1|X)$ esta en el rango 

```python
# Gráfico de resultados
plt.figure(figsize=(4,2))

# Ordenamos los valores de X
X_sorted=np.sort(X, axis=0).flatten()
# Predicciones ordenadas (probabilidad)
y_pred_score_sorted = log_reg.predict_proba(X_sorted.reshape(-1,1))[:,1]   

plt.plot(X_sorted, y_pred_score_sorted, color='blue', zorder=10, label='Prob predicha')
plt.scatter(X, y, color='orange', zorder=20, marker="|", label='Y obs.' )
plt.xlabel('x')
plt.ylabel('Prob(y=1|X)')
plt.legend()
plt.show()
```

```python
print(y_pred_score_sorted)
```

Ahora, para la predicción de $\hat{Y}=1$ o $\hat{Y}=0$ usando el modelo estimado de regresión logística, debemos usar la regla de decisión de Bayes de $\hat{p}>0.5$  

```python
# Convertimos las probabilidades en Y con valores 1 o 0 (usando el clasificador de Bayes)
y_pred = np.where(y_pred_score > 0.5, 1, y_pred_score)
y_pred = np.where(y_pred_score <= 0.5, 0, y_pred)
print('Proporcion de 1s predichos: ', y_pred.mean().round(3))
print('Proporcion de 1s reales: ', y.mean().round(3))
```

```python
# Equivalente a lo anterior es usar predict() (clasifica en 0s y 1s usando el umbral 0.5)
y_pred_2 = log_reg.predict(X)
print(pd.crosstab(index=y_pred, columns=y_pred_2)) #tabla para chequear que la prediccion es igual

# Y graficamos los resultados
plt.figure(figsize=(6, 4))
plt.scatter(X, y_pred_2, color='red', alpha=0.3, zorder=20, label='Y predicho')
plt.scatter(X, y, color='green', zorder=20, marker="|", label='Y obs.')
plt.legend()
plt.show()
```

#### REGRESIÓN LOGÍSTICA CON STATSMODELS

```python
# Podemos repetirlo con statsmodels
# Primero agregamos la columna de 1s y hacemos el ajuste
X_sm = sm.add_constant(X) 
logit_model = sm.Logit(y, X_sm)
result = logit_model.fit()
print(result.summary2()) 
#También podríamos exportarla como latex
print(result.summary2().as_latex())
```

Ahora, exportamos en una tabla, los coeficientes y ratio de probabilidades

```python
# Guardamos los coeficientes
coeficientes = log_reg.coef_[0]

# Calculamos los odds ratio
odds_ratio = np.exp(coeficientes) #La regresión logística estima los coeficientes en escala logarítmica (log-odds). Para volver a una escala interpretable, se aplica exponencial

# Guardamos el error estándar
std_error = result.bse[1:]  # Excluir el intercepto

# Creamos la tabla
tabla_resultados = pd.DataFrame({
    #'Variable': X.columns esta linea para cuando tengan muchas variables
    'Coeficiente': coeficientes,
    'Std. Error': std_error,
    'Odds Ratio': odds_ratio
})

tabla_resultados.round(2)
#Cuando X sube una unidad, la probabilidad de que Y sea 1 se multiplican por .....
```

```python
y_pred_score_sm = result.predict(X_sm)

# El método where requiere una condición como primer parámetro, 
# que cuando es True devuelve el segundo valor y cuando es False devuelve tercero. 
y_pred_sm = np.where(y_pred_score_sm > 0.5, 1, y_pred_score_sm)
y_pred_sm = np.where(y_pred_score_sm <= 0.5, 0, y_pred_sm)

print(pd.crosstab(index=y_pred, columns=y_pred_sm))
```

### Medidas de desempeño del modelo de clasifiación 

Dependiendo la prioridad del problema seguramente vamos a querer usar diferentes métricas. Scikit learn tiene muchas métricas que pueden explorar en el módulo [metrics](https://scikit-learn.org/stable/modules/model_evaluation.html)

- Sensitivity o Recall o True Positive Rate: TP rate = TP/P
- Specificity o True Negative Rate: 1 - FP rate = TN/N
- False Positive Rate o False Alarm Rate: FP rate = FP/N
- False Negative Rate: FN rate = FN/P
- Precision o Positive Predicted Value: TP/(TP+FP)
- Accuracy: (TP+TN)/(P+N)

Nota: Cuidado con las traducciones! "Accuracy" lo pueden encontrar traducido como "precisión" y eso puede generar confusión con la medida "precision" (o positive predicted value). Mi sugerencia es traducir "accuracy" como "exactitud".

| | **Predicted A** | **Predicted B** | **Predicted C** |
| :--- | :---: | :---: | :---: |
| **Actual A** | $TP_A$ | $FN_{AB}$ | $FN_{AC}$ |
| **Actual B** | $FP_{BA}$ | $TP_B$ | $FN_{BC}$ |
| **Actual C** | $FP_{CA}$ | $FP_{CB}$ | $TP_{C}$ |

Donde $FP_{ij}$ es falso positivo de la fila $i$ columna $j$, $FN_{ij}$ es falso negativo de la fila $i$ columna $j$ y $TP_i$ es el verdadero positivo de la fila $i$

## Análisis Discriminante Lineal (LDA)

El **LDA** es otro clasificador clásico que asume que cada clase sigue una distribución normal con la misma matriz de covarianza.

**Intuición geométrica:** LDA encuentra el eje sobre el cual las clases están más separadas entre sí (gran distancia entre medias) y más compactas dentro de cada clase (baja varianza interna). La frontera de decisión resultante es **lineal**.
Cuando los datos dentro de cada clase son aproximadamente normales y tenés pocas observaciones, LDA es más eficiente. Con muchos datos, la diferencia desaparece.

Lo vemos con **datos ficticios** para que la idea quede clara.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from matplotlib.colors import ListedColormap
```

```python
np.random.seed(42)
N = 100  # obs. por clase

# Clase 0: media (2, 2),  Clase 1: media (6, 6)
X_c0 = np.random.multivariate_normal(mean=[2, 2], cov=[[1.2, 0.5],[0.5, 1.0]], size=N)
X_c1 = np.random.multivariate_normal(mean=[6, 6], cov=[[1.0, 0.3],[0.3, 1.2]], size=N)

X_lda = np.vstack([X_c0, X_c1])
y_lda = np.array([0]*N + [1]*N)

print(f"Observaciones totales: {len(y_lda)}")
print(f"  Clase 0 (ej. 'Down'): {(y_lda==0).sum()}")
print(f"  Clase 1 (ej. 'Up'):   {(y_lda==1).sum()}")
```

```python
# Entrenamos el LDA
lda = LinearDiscriminantAnalysis()
lda.fit(X_lda, y_lda)

print("Medias estimadas por clase:")
print(f"  Clase 0: X1={lda.means_[0][0]:.2f}, X2={lda.means_[0][1]:.2f}")
print(f"  Clase 1: X1={lda.means_[1][0]:.2f}, X2={lda.means_[1][1]:.2f}")
```

```python
# ---- Visualización: frontera + proyección 1D ----
xx, yy = np.meshgrid(np.linspace(-1, 10, 300), np.linspace(-1, 10, 300))
Z_lda  = lda.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

cmap_bg = ListedColormap(['red', 'green'])
rng = np.random.default_rng(1)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Izquierda: frontera de decisión en 2D
axes[0].contourf(xx, yy, Z_lda, alpha=0.35, cmap=cmap_bg)
axes[0].scatter(X_c0[:,0], X_c0[:,1], c='red',  label='Clase 0', edgecolors='k', s=40, zorder=2)
axes[0].scatter(X_c1[:,0], X_c1[:,1], c='blue', label='Clase 1', edgecolors='k', s=40, zorder=2)
axes[0].set_xlabel('Variable X₁', fontsize=11)
axes[0].set_ylabel('Variable X₂', fontsize=11)
axes[0].set_title('LDA — Frontera de decisión (lineal)', fontsize=12)
axes[0].legend()

# Derecha: proyección sobre el eje discriminante (1D)
X_proj = lda.transform(X_lda).flatten()
jitter = rng.uniform(-0.08, 0.08, size=len(y_lda))
axes[1].scatter(X_proj[y_lda==0], jitter[y_lda==0], c='red',  alpha=0.5, label='Clase 0', s=30)
axes[1].scatter(X_proj[y_lda==1], jitter[y_lda==1], c='blue', alpha=0.5, label='Clase 1', s=30)
axes[1].axvline(0, color='black', lw=2, linestyle='--', label='Frontera (umbral = 0)')
axes[1].set_xlabel('Proyección sobre el eje LDA', fontsize=11)
axes[1].set_yticks([])
axes[1].set_title('Proyección 1D — separación entre clases', fontsize=12)
axes[1].legend()

plt.suptitle('LDA con datos ficticios — la frontera es una línea recta', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

#### Evaluación de la performance del LDA fuera de la muestra

```python
# Train-test split
from sklearn.model_selection import train_test_split as tts_lda

X_tr_lda, X_te_lda, y_tr_lda, y_te_lda = tts_lda(X_lda, y_lda, test_size=0.3, random_state=42)

lda_oos = LinearDiscriminantAnalysis()
lda_oos.fit(X_tr_lda, y_tr_lda)
y_pred_lda = lda_oos.predict(X_te_lda)

print("=== PERFORMANCE DEL LDA (test set) ===")
print(f"Accuracy : {accuracy_score(y_te_lda, y_pred_lda):.4f}")
print(f"Recall   : {recall_score(y_te_lda, y_pred_lda):.4f}")
print("\nMatriz de confusión:")
print(confusion_matrix(y_te_lda, y_pred_lda))
```

#### ¿Cuándo usar LDA en vez de Logit?

| Aspecto | Logit | LDA |
|---------|-------|-----|
| Supuesto sobre X | Ninguno | X ~ Normal por clase |
| Frontera de decisión | Lineal | Lineal |
| Ventaja principal | Robusto a no-normalidad | Más eficiente si X es normal |
| Multi-clase | Logit multinomial | LDA  |

**Regla práctica:** con datos que se parecen a normales y pocas observaciones, LDA suele funcionar tan bien o mejor que Logit. Con muchos datos, la diferencia tiende a desaparecer.
