##### Funciones: DocString, repaso
Cómo podemos repetir el código sin escribirlo por tercera vez?

Podemos hacer que nuestro código funcione para otros grados?

```python
def transf_reg_poly(grado, x_train, x_test, y_train, y_test):
    '''
    La función realiza una transformación polinomial y luego corre una regresión lineal polinómica
    Input:
        grado
        x_train, x_test, y_train, y_test
    Output:
        modelo, mse
    '''
    poly = PolynomialFeatures(degree = grado, include_bias=False) 

    x_train_poly = poly.fit_transform(x_train)
    x_test_poly = poly.fit_transform(x_test)  
  
    model = LinearRegression().fit(x_train_poly, y_train) 
    y_pred_poly = model.predict(x_test_poly)
    
    mse = mean_squared_error(y_test, y_pred_poly)
    return model, mse
```

```python
mse1 = transf_reg_poly(1, x_train, x_test, y_train, y_test)[1]
mse2b = transf_reg_poly(2, x_train, x_test, y_train, y_test)[1]
mse3b = transf_reg_poly(3, x_train, x_test, y_train, y_test)[1]
mse4b = transf_reg_poly(4, x_train, x_test, y_train, y_test)[1]
mse5b = transf_reg_poly(5, x_train, x_test, y_train, y_test)[1]
```

```python
# Performance con la PRIMERA partición de muestras de entrenamiento y testeo
print('Regresión Grado 1:', mse_lreg)
print('Regresión Grado 2:', mse2)
print('Regresión Grado 3:', mse3b)
```

```python
# Performance con la SEGUNDA partición de muestras de entrenamiento y testeo
print('Regresión lineal:', mse1)
print('Regresión Grado 2:', mse2b)
print('Regresión Grado 3:', mse3b)
print('Regresión Grado 4:', mse4b)
print('Regresión Grado 5:', mse5b)
```

###  2. K-FOLD CROSS-VALIDATION  

Es un **técnica de remuestreo**. Se usa para estimar el error (test) asociado a un método de aprendizaje, para:  
- Elegir el nivel de complejidad optimo (Model selection)
- Evaluar el error de pronóstico fuera de la muestra (futura, condicional, contra fáctica, etc.) (Model Assesment)

Consiste en:
- Dividir las observaciones en k folds (pliegues), del mismo tamaño, aleatoriamente. 
- Ajustar el modelo k veces, cada vez con k-1 folds (distintos cada vez). Computar k veces el error de predicción en el fold reservado. (cada fold se usa k-1 veces como training set y 1 vez como test set).
- Estimar el error de predicción, estimación que surge de promediar las K estimaciones obtenidas.

Vamos a usar [KFold](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html) de Scikitlearn

## Breve mencion al problema de la _"Maldicion de la Multidimensionalidad"_

Hasta aqui probamos un modelo no lineal (*regresion polinomica*) con un solo predictor (`horsepower`).

**Pregunta**: Qué pasa si tenemos 2, 3, o más predictores?

```python
auto_d.info(verbose=True)
```

```python
# Creamos una funcion que nos vaya creando una matriz X con 1, 2, ... 9 predictores.
def crear_matrices_X(auto_d):
    matrices_X = {}
    for i in range(1, 11):
        columnas = auto_d.columns[1:i+1]  # Selecciona las columnas desde 'cylinders' hasta la i-ésima columna
        X_i = auto_d[columnas].values  # Crea la matriz X_i
        matrices_X[f'X{i}'] = X_i  # Almacena la matriz X_i en un diccionario
    return matrices_X
```

```python
# Creamos las matrices
matrices_X = crear_matrices_X(auto_d)
```

```python
X2 = matrices_X['X2']
# Veamos la matriz con 2 predictores si quisieramos hacer un modelo cuadrático:
poly = PolynomialFeatures(degree = 2, include_bias=False) 

# Transformamos el vector columna en una matriz para tener en cuenta el grado del polinomio de interés
print('X antes de la transformación:\n', X2.shape)
X2_poly = poly.fit_transform(X2)
np.set_printoptions(suppress = True) # evita que el print salga con notación científica
print('X luego de la transformación:\n', X2_poly.shape)
```

Ahora generalicemos esta idea, para responder la pregunta de la dimensionalidad

```python
def crear_polinomios(matrices_X, d):
    resultados = {}
    for nombre, X in matrices_X.items():
        poly_features = PolynomialFeatures(degree=d, include_bias=False)
        X_poly = poly_features.fit_transform(X)
        num_columnas = X_poly.shape[1]
        resultados[nombre] = num_columnas
    return resultados
```

```python
d = 2  # Por ejemplo, grado 2
resultados = crear_polinomios(matrices_X, d)
print(resultados)
```

```python
d = 3  # Por ejemplo, grado 3
resultados = crear_polinomios(matrices_X, d)
print(resultados)
```

```python
d = 4  # Por ejemplo, grado 3
resultados = crear_polinomios(matrices_X, d)
print(resultados)
```

En el caso del _modelo no lineal_ de regresion polinómica cae rapidamente en el problema de la dimensionalidad. Aqui con $n=392$ no podriamos estimar un modelo tan complejo con 8 predictores y grado del polinomio 4, ya que no tenemos suficiente grados de libertad.

Esta breve mención e ilustracion nos muestran que la **seleccion de complejidad** del modelo por cross-validation está restringuida por el tamaño de la muestra.
